```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load a grayscale image
image_path = 'emma.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define the stepwise function for intensity transformation
def stepwise_intensity_transform(pval):
    if pval < 50:
        return pval
    elif pval < 150:
        return 1.55*pval+22.5
    else:
        return pval

# Apply the stepwise intensity transformation to the image
transformed_image = np.vectorize(stepwise_intensity_transform)(original_image)

# Convert the transformed image to the appropriate data type (uint8)
transformed_image = transformed_image.astype(np.uint8)
fig, axes = plt.subplots(1, 2, figsize=(18, 9))
axes[0].imshow(original_image, cmap='gray')
axes[0].set_title('Original')
axes[0].set_xticks([]), axes[0].set_yticks([])

axes[1].imshow(transformed_image, cmap='gray')
axes[1].set_title('Transformed')
axes[1].set_xticks([]), axes[1].set_yticks([])
```
